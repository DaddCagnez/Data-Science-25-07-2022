## Lavori Attivati

```python
# Import the dataframe 'Rapporti_di_lavoro_attivati.csv'
lavori_attivati = pd.read_csv('Rapporti_di_lavoro_attivati.csv')
#lavori_attivati.head()
```

```python
lavori_attivati
```

```python
print(lavori_attivati.isnull().sum())  # see if there are any missing values
```

```python
# Removal null values from the table
cols = ['SETTOREECONOMICODETTAGLIO','TITOLOSTUDIO', 'MODALITALAVORO']
for column in cols:
    lavori_attivati.loc[lavori_attivati[column] == '?', column] = np.nan
    lavori_attivati.dropna(axis = 0, inplace = True)
    lavori_attivati.shape
print(lavori_attivati.isnull().sum())  # see if there are any missing values
```

```python
# Extraction Year from 'DATA' column
lavori_attivati['DATA']= pd.to_datetime(lavori_attivati['DATA']) 
lavori_attivati['YEAR'] = lavori_attivati['DATA'].dt.year
new_lavori_attivati=lavori_attivati.drop(["DATA"], axis=1)
new_lavori_attivati =  new_lavori_attivati[new_lavori_attivati['YEAR'] == 2021]
new_lavori_attivati["TIPO"]="ATTIVATO"
new_lavori_attivati.head(5)
```

## Lavori Cessati

```python
# Import the dataframe 'Rapporti_di_lavoro_cessati.csv'
lavori_cessati = pd.read_csv('Rapporti_di_lavoro_cessati.csv')
lavori_cessati.head(1)
```

```python
print(lavori_cessati.isnull().sum()) #see if there are any missing values
```

```python
# Removal null values from the table
cols = ['SETTOREECONOMICODETTAGLIO','TITOLOSTUDIO', 'MODALITALAVORO']
for column in cols:
    lavori_cessati.loc[lavori_cessati[column] == '?', column] = np.nan
    lavori_cessati.dropna(axis = 0, inplace = True)
    lavori_cessati.shape
print(lavori_cessati.isnull().sum()) #see if there are any missing values
```

```python
# Extraction Year from 'DATA' column
lavori_cessati['DATA']= pd.to_datetime(lavori_cessati['DATA'], errors = 'coerce')
lavori_cessati['YEAR'] = lavori_cessati['DATA'].dt.year
new_lavori_cessati = lavori_cessati.drop(['DATA'], axis=1)
new_lavori_cessati =  new_lavori_cessati[new_lavori_cessati['YEAR'] == 2021]
new_lavori_cessati["TIPO"] = "CESSATO"
new_lavori_cessati.head(5)
```

```python
# Removal of floating point from 'YEAR' column values
d = new_lavori_cessati["YEAR"].size
for i in range(d):
    c=int(new_lavori_cessati["YEAR"].values[i])
    new_lavori_cessati["YEAR"] = c
new_lavori_cessati.head()
```
